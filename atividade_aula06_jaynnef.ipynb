```python
import pandas as pd
```

```python
df = pd.read_csv('https://storage.googleapis.com/atividade-aula06-jaynnef/dados-brutos/dados_inconsistentes%20-%20dados_inconsistentes.csv.csv')
print(df)
```

```python
df.info()
```

```python
# lista com as colunas
a = len(df)
i = 0
nome = list(df['nome'])
idade = list(df['idade'])
salario = list(df['salário'])
cargo = list(df['cargo'])
departamento = list(df['departamento'])
display(df)
```

```python
a = len(df)
i = 0

# Atribuindo 0 aos valores nulos

for i in range(a):

  if type(nome[i]) == float:
    nome[i] = 0

  if idade[i] > 0:
    idade[i] = idade[i]
  else:
    idade[i] = 0

  if type(salario[i]) == float:
    salario[i] = 0

  if type(cargo[i]) == float:
    cargo[i] = 0

  if type(departamento[i]) == float:
    departamento[i] = 0

# idade, salario, cargo e departamento = NaN, salario = erro ou departamento = Desconhecido
while i < len(nome):
  if idade[i] != idade[i] or salario[i] != salario[i] or cargo[i] != cargo[i] or departamento[i] != departamento[i] or salario[i] == 'erro' or departamento[i] == 'Desconhecido':
    del df[i]
    i = 0
  else:
    i = i + 1
```

```python
#dados tratados inserindo em nova lista
nome_nv = list()
idade_nv = list()
salario_nv = list()
cargo_nv = list()
departamento_nv = list()

for i in range(a):
  if nome[i] != 0 and idade[i] != 0 and salario[i] != 0 and cargo[i] != 0 and departamento[i] != 0:
    nome_nv.append(nome[i])
    idade_nv.append(idade[i])
    salario_nv.append(salario[i])
    cargo_nv.append(cargo[i])
    departamento_nv.append(departamento[i])

```

```python
df = pd.DataFrame(index = range(len(nome_nv)), columns=['nome','idade','salario','cargo','departamento'])

df['nome'] = nome_nv
df['idade'] = idade_nv
df['salario'] = salario_nv
df['cargo'] = cargo_nv
df['departamento'] = departamento_nv

```

```python
print(df)
```

```python
df.to_csv('tratados.csv',index=False)
```
